```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
```

```python
class Student(Person):
    def __init__(self, name, age,student_id,enroll_course):
        super().__init__(name, age)
        self.student_id = student_id
        self.enroll_course = enroll_course
    def display_info(self):
        super().display_info()
        print("Student ID:", self.student_id)
        print("Enroll Course:", self.enroll_course)
```

```python
class Lecturer(Person):
    def __init__(self, name, age,employee_id,teach_course):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.teach_course = teach_course
    def display_info(self):
        super().display_info()
        print("Employee ID:", self.employee_id)
        print("Teaching Course:", self.teach_course)
```

```python
student1=Student("Maxwell Mensah",20,"FOE.41,006.110.25","Object Oriented Programming")
lecturer1=Lecturer("Mr.Mathew Cobbinah",42,"UMaT-LT-025","Objected Oriented Programming")
print("------STUDENT------")
student1.display_info()
print("------LECTURER------")
lecturer1.display_info()
```
